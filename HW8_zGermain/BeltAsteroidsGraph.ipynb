```
ax = pl.figure(figsize = [11,8.5])
pl.scatter(ast["a"], ast["e"], s = 0.5, c = ast["albedo"], cmap="cool", alpha = 0.5)
pl.colorbar(label="Geometric Albedo")
pl.xlabel("Semi-Major Axis (AU)")
pl.ylabel("Eccentricity")
#Belt seperators
pl.axvline(x = 2, ymax = 0.05, c = 'red') #Edge of inner belt
pl.axvline(x = 3.2, ymax = 0.05, c = 'red') #Edge of Main Belt
pl.axvline(x = 4.6, ymax = 0.05, c = 'red') #Edge of Outer Belt
#Kirkwood Gaps
pl.axvline(x = 2.06, ymax = 0.05, c = 'orange') #4:1 Resonance
pl.axvline(x = 2.5, ymax = 0.05, c = 'orange') #3:1
pl.axvline(x = 2.82, ymax = 0.05, c = 'orange') #5:2
pl.axvline(x = 3.28, ymax = 0.05, c = 'orange'); #2:1
```
